Season-by-Season Win Trajectory of Mumbai Indians from 2008 to 2024.



 Step 1: Filter MI Matches

```python
#  Import libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
%matplotlib inline

# 📥 Load the datasets
matches = pd.read_csv('../data/matches.csv')
deliveries = pd.read_csv('../data/deliveries.csv')
```

```python
# Filter matches where MI participated
mi_matches = matches[(matches['team1'] == 'Mumbai Indians') | (matches['team2'] == 'Mumbai Indians')].copy()
```

```python
# Add a column to indicate if MI won that match
mi_matches['mi_win'] = mi_matches['winner'] == 'Mumbai Indians'
```

```python
# Total matches played per season
matches_per_season = mi_matches.groupby('season').size()

# Total wins per season
wins_per_season = mi_matches.groupby('season')['mi_win'].sum()

# Combine both into a dataframe
mi_season_stats = pd.DataFrame({
    'Matches Played': matches_per_season,
    'Matches Won': wins_per_season,
    'Win Percentage': (wins_per_season / matches_per_season * 100).round(2)
})
print(mi_season_stats)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(12, 6))
sns.barplot(x=mi_season_stats.index, y=mi_season_stats['Matches Won'], palette='Blues_d')
plt.title('🏆 Mumbai Indians - Matches Won per Season (2008–2024)')
plt.xlabel('Season')
plt.ylabel('Matches Won')
plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(12, 5))
sns.lineplot(data=mi_season_stats, x=mi_season_stats.index, y='Win Percentage', marker='o', color='navy')
plt.title('MI Win Percentage by Season (2008–2024)')
plt.xlabel('Season')
plt.ylabel('Win Percentage')
plt.ylim(0, 100)
plt.grid(True, linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
```
