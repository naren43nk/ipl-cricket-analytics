## 📊 IPL Analytics 2008–2024 - EDA Notebook

### This notebook explores the IPL dataset including match-level and ball-by-ball data.
 - `matches.csv`: Contains high-level info of each IPL match.
 - `deliveries.csv`: Contains ball-by-ball commentary data.

### Objectives:
### - Load and inspect the data
### - Clean column names and formats
### - Explore trends in wins, tosses, venues, players, and teams


```python
#  Import libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
%matplotlib inline
```

```python
import os
print("Current working directory:", os.getcwd())
```

```python
# 📥 Load the datasets
matches = pd.read_csv('../data/matches.csv')
deliveries = pd.read_csv('../data/deliveries.csv')
```

### Data Cleaning Plan

```python
#  Peek at match-level data
matches.head()
```

### 1.1 Matches.csv Cleaning Checklist

```python
# Basic info
matches.info()
matches.isnull().sum()
```

```python
matches['date'] = pd.to_datetime(matches['date'], errors='coerce')
matches['team1'] = matches['team1'].str.strip()
matches['team2'] = matches['team2'].str.strip()
matches['winner'] = matches['winner'].fillna('No Result')
```

### 1.2 deliveries.csv Cleaning Checklist

```python
deliveries = pd.read_csv('../data/deliveries.csv')
deliveries.info()
deliveries.isnull().sum()
```

```python
deliveries['extras_type'] = deliveries['extras_type'].fillna('None')
deliveries['player_dismissed'] = deliveries['player_dismissed'].fillna('None')
```

### Exploratory Data Analysis (EDA)
### 2.1 Basic Match Insights

```python
print("Seasons:", matches['season'].nunique())
print("Teams:", pd.unique(matches[['team1','team2']].values.ravel()))
print("Venues:", matches['venue'].nunique())
```

### 2.2 Matches per Season

```python
matches_per_season = matches['season'].value_counts().sort_index()
matches_per_season.plot(kind='bar', title='📅 Number of Matches per Season')
```

### 2.3 Toss vs Match Wins



```python
matches['toss_match_win'] = matches['toss_winner'] == matches['winner']
toss_match_win_pct = matches['toss_match_win'].mean() * 100
print(f"Winning after toss win: {toss_match_win_pct:.2f}%")
```

### 2.4 Team Win Distribution

```python
win_counts = matches['winner'].value_counts().sort_values(ascending=True)
win_counts.plot(kind='barh', figsize=(8, 10), title=' Team Win Counts')
```

### 2.5 Batsmen & Bowlers Overview (Whole League)

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Group by batter and sum runs
top_run_scorers = deliveries.groupby('batter')['batsman_runs'].sum().sort_values(ascending=False).head(10)

# Display as a table
print("🏏 Top 10 Run Scorers:")
print(top_run_scorers)

# Plot bar chart
plt.figure(figsize=(10,6))
sns.barplot(x=top_run_scorers.values, y=top_run_scorers.index, palette="viridis")
plt.title("🏏 Top 10 IPL Run Scorers (2008–2024)")
plt.xlabel("Total Runs")
plt.ylabel("Batter")
plt.grid(axis='x', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
# Filter out dismissals that are not wickets (e.g., run out)
dismissals = deliveries[deliveries['dismissal_kind'].notna()]
dismissals = dismissals[dismissals['dismissal_kind'] != 'run out']

# Group by bowler
top_wicket_takers = dismissals.groupby('bowler').size().sort_values(ascending=False).head(10)

# Display
print(" Top 10 Wicket Takers:")
print(top_wicket_takers)

# Plot bar chart
plt.figure(figsize=(10,6))
sns.barplot(x=top_wicket_takers.values, y=top_wicket_takers.index, palette="magma")
plt.title(" Top 10 IPL Wicket Takers (2008–2024)")
plt.xlabel("Total Wickets")
plt.ylabel("Bowler")
plt.grid(axis='x', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```
