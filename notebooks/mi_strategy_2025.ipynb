### üü¶ Mumbai Indians IPL 2025 Strategy Notebook
This notebook analyzes IPL match data from 2008 to 2024 to uncover tactical insights 
for the Mumbai Indians and offers data-driven recommendations for IPL 2025.



```python
# Load match and ball-by-ball datasets
import pandas as pd

matches = pd.read_csv('../data/matches.csv')
deliveries = pd.read_csv('../data/deliveries.csv')
```

### Season-by-Season Performance Summary

```python
# Filter only MI matches
mi_matches = matches[(matches['team1'] == 'Mumbai Indians') | (matches['team2'] == 'Mumbai Indians')].copy()

# Add a win flag for MI
mi_matches['mi_win'] = mi_matches['winner'] == 'Mumbai Indians'
```

```python
# Group by season and calculate wins, matches, and win percentage
season_stats = mi_matches.groupby('season')['mi_win'].agg(['sum', 'count'])
season_stats.columns = ['Matches Won', 'Matches Played']
season_stats['Win %'] = (season_stats['Matches Won'] / season_stats['Matches Played'] * 100).round(2)
season_stats
```

### Classify Seasons as Peak / Slump / Average

```python
# Classify performance levels
def classify_performance(win_pct):
    if win_pct >= 60:
        return 'üèÜ Peak'
    elif win_pct < 40:
        return 'üìâ Slump'
    else:
        return '‚öñÔ∏è Average'

season_stats['Performance'] = season_stats['Win %'].apply(classify_performance)
season_stats
```

### Visualize Season-Wise Wins



```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(12, 6))
sns.barplot(x=season_stats.index, y='Matches Won', data=season_stats, palette='Blues_d')
plt.title('üèÜ Mumbai Indians - Matches Won per Season (2008‚Äì2024)')
plt.xlabel('Season')
plt.ylabel('Wins')
plt.grid(axis='y', linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
```

### Win % by Season + Performance Level


```python
plt.figure(figsize=(12, 6))
sns.barplot(x=season_stats.index, y='Win %', hue='Performance', data=season_stats, palette={
    'üèÜ Peak': 'royalblue', 'üìâ Slump': 'crimson', '‚öñÔ∏è Average': 'gold'
})
plt.title(' MI Season Performance: Win % and Classification')
plt.xlabel('Season')
plt.ylabel('Win Percentage')
plt.grid(axis='y', linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
```

### 1:Winning Pattern Detection (Toss, Batting First/Second)
### 1.1 Add Toss & Inning Info for MI Matches

```python
# Re-filter MI matches to include toss and innings info
mi_matches = matches[(matches['team1'] == 'Mumbai Indians') | (matches['team2'] == 'Mumbai Indians')].copy()
mi_matches['mi_win'] = mi_matches['winner'] == 'Mumbai Indians'

# Was toss won by MI?
mi_matches['toss_win'] = mi_matches['toss_winner'] == 'Mumbai Indians'

# Was MI batting first?
# Mumbai Indians = team1 and toss_decision = 'bat' ‚Üí batted first
def determine_batting_first(row):
    if row['toss_winner'] == 'Mumbai Indians':
        return row['toss_decision'] == 'bat'
    else:
        return row['toss_decision'] == 'field'

mi_matches['batting_first'] = mi_matches.apply(determine_batting_first, axis=1)
```

### 1.2 Toss vs Match Win Correlation


```python
# MI win percentage when they won the toss
toss_win_stats = mi_matches.groupby('toss_win')['mi_win'].mean().round(2) * 100
print("Toss Win vs Match Win %:\n", toss_win_stats)
```

### 1.3 Batting First vs Second Success


```python
# MI win % when batting first or second
batting_first_stats = mi_matches.groupby('batting_first')['mi_win'].mean().round(2) * 100
print(" Batting First vs Second Win %:\n", batting_first_stats)
```

### 1.4 Visualization

### 2: Top Batsmen Breakdown (Mumbai Indians)

```python
# Filter deliveries where MI is the batting team
mi_batting = deliveries[deliveries['batting_team'] == 'Mumbai Indians']
```

```python
# Total runs scored
mi_batsmen_runs = mi_batting.groupby('batter')['batsman_runs'].sum().sort_values(ascending=False)
top_batsmen = mi_batsmen_runs.head(10)
print("üèè Top 10 MI Run Scorers:\n", top_batsmen)
```

### 3: Playoff Stage Classification

### 3.1 Sort matches by season and date
We‚Äôll use the date column instead of match_id to identify playoffs by sorting matches within each season.

```python
# Ensure 'date' is datetime if not already
matches['date'] = pd.to_datetime(matches['date'], errors='coerce')

# Sort matches by season and date
matches_sorted = matches.sort_values(['season', 'date']).copy()

# Add sequential number to each match within a season
matches_sorted['match_order'] = matches_sorted.groupby('season').cumcount() + 1

# Total matches in each season
matches_sorted['total_matches'] = matches_sorted.groupby('season')['season'].transform('count')
```

### 3.2 Mark Last 4 Matches as Playoffs

```python
# Last 4 matches of each season are usually: Qualifier 1, Eliminator, Qualifier 2, Final
matches_sorted['is_playoff'] = matches_sorted['match_order'] >= (matches_sorted['total_matches'] - 3)
```

### 3.3 Merge Playoff Info into MI Matches

```python
# Filter MI matches and merge playoff flag
mi_matches = matches_sorted[(matches_sorted['team1'] == 'Mumbai Indians') | (matches_sorted['team2'] == 'Mumbai Indians')].copy()
mi_matches['mi_win'] = mi_matches['winner'] == 'Mumbai Indians'
```

### 3.4 View MI Playoff History

```python
# View MI playoff matches
mi_playoffs = mi_matches[mi_matches['is_playoff'] == True][['season', 'date', 'winner', 'venue']]
print("üìÖ MI Playoff Matches:")
print(mi_playoffs.sort_values('season'))
```

### 3.5 Add Stage Labels

```python
def label_stage(row):
    order = row['match_order']
    total = row['total_matches']
    if order == total:
        return 'Final'
    elif order == total - 1:
        return 'Qualifier 2'
    elif order == total - 2:
        return 'Eliminator'
    else:
        return 'Qualifier 1'

matches_sorted['stage'] = matches_sorted.apply(lambda row: label_stage(row) if row['is_playoff'] else 'League', axis=1)

# Re-merge with MI
mi_matches = matches_sorted[(matches_sorted['team1'] == 'Mumbai Indians') | (matches_sorted['team2'] == 'Mumbai Indians')].copy()
```

### 4: Bowling Phase Performance Analysis


### 4.1 Filter MI Bowling Data

```python
# Filter only Mumbai Indians bowling deliveries
mi_bowling = deliveries[deliveries['bowling_team'] == 'Mumbai Indians'].copy()
```

### 4.2 Add Phase Label

```python
# Create bowling phase label based on over
def classify_phase(over):
    if over <= 6:
        return 'Powerplay'
    elif 7 <= over <= 15:
        return 'Middle'
    else:
        return 'Death'

mi_bowling['phase'] = mi_bowling['over'].apply(classify_phase)
```

### 4.3 Wickets per Phase

```python
# Consider only actual dismissals (not runouts, etc.)
valid_dismissals = ['caught', 'bowled', 'lbw', 'stumped', 'caught and bowled', 'hit wicket']
mi_bowling_wickets = mi_bowling[mi_bowling['dismissal_kind'].isin(valid_dismissals)]

# Count wickets per phase
wickets_by_phase = mi_bowling_wickets.groupby('phase').size()
print("üß® Wickets by Phase:\n", wickets_by_phase)
```

### 4.4 Economy Rate per Phase

```python
# Total runs conceded per phase
runs_by_phase = mi_bowling.groupby('phase')['total_runs'].sum()

# Total legal deliveries (no wides or no-balls)
legal_deliveries = mi_bowling[(mi_bowling['extras_type'] != 'wides') & 
                              (mi_bowling['extras_type'] != 'noballs')].groupby('phase').size()

# Calculate overs
overs_by_phase = legal_deliveries / 6

# Economy rate = total runs / overs
economy_by_phase = (runs_by_phase / overs_by_phase).round(2)
print("üí∏ Economy Rate by Phase:\n", economy_by_phase)
```

### 4.5 Plot Wickets & Economy per Phase


```python
import matplotlib.pyplot as plt

# Combined plot
fig, ax1 = plt.subplots(figsize=(8, 5))

color = 'tab:red'
ax1.set_title(' MI Bowling Phase Performance (2008‚Äì2024)')
ax1.set_ylabel('Wickets', color=color)
ax1.bar(wickets_by_phase.index, wickets_by_phase.values, color=color, alpha=0.6)
ax1.tick_params(axis='y', labelcolor=color)

# Twin axis for economy rate
ax2 = ax1.twinx()
color = 'tab:blue'
ax2.plot(economy_by_phase.index, economy_by_phase.values, color=color, marker='o', linewidth=2)
ax2.set_ylabel('Economy Rate', color=color)
ax2.tick_params(axis='y', labelcolor=color)

plt.tight_layout()
plt.grid(True, linestyle='--', alpha=0.4)
plt.show()
```

### 5: Bowler-Wise Performance by Phase
### 5.1 Prepare Phase Labels


```python
# Reuse if already done; otherwise apply
def classify_phase(over):
    if over <= 6:
        return 'Powerplay'
    elif 7 <= over <= 15:
        return 'Middle'
    else:
        return 'Death'

mi_bowling['phase'] = mi_bowling['over'].apply(classify_phase)
```

### 5.2 Calculate Wickets per Bowler per Phase


```python
# Filter valid dismissals
valid_dismissals = ['caught', 'bowled', 'lbw', 'stumped', 'caught and bowled', 'hit wicket']
mi_wickets = mi_bowling[mi_bowling['dismissal_kind'].isin(valid_dismissals)]

# Group: bowler + phase ‚Üí wickets
bowler_phase_wickets = mi_wickets.groupby(['bowler', 'phase']).size().unstack(fill_value=0)
bowler_phase_wickets.head()
```

### 5.3 Economy Rate per Bowler per Phase

```python
# Runs conceded per bowler-phase
runs_conceded = mi_bowling.groupby(['bowler', 'phase'])['total_runs'].sum()

# Legal deliveries per bowler-phase
legal_deliveries = mi_bowling[~mi_bowling['extras_type'].isin(['wides', 'noballs'])] \
                    .groupby(['bowler', 'phase']).size()

# Overs
overs_bowled = legal_deliveries / 6

# Economy = runs / overs
economy_rate = (runs_conceded / overs_bowled).round(2)
economy_rate = economy_rate.unstack(fill_value=0)
economy_rate.head()
```

### 5.4 Focus on Death Overs: Top Death Bowlers (min 30 overs)


```python
#  Step 1: Filter to only Death overs
death_bowling = mi_bowling[mi_bowling['phase'] == 'Death']

#  Step 2: Legal deliveries (exclude wides & no-balls)
death_legal = death_bowling[~death_bowling['extras_type'].isin(['wides', 'noballs'])]

#  Step 3: Aggregate runs and balls bowled
death_runs = death_bowling.groupby('bowler')['total_runs'].sum()
death_balls = death_legal.groupby('bowler').size()

#  Step 4: Filter bowlers with minimum 30 overs = 180 balls
eligible_bowlers = death_balls[death_balls > 180].index

#  Step 5: Calculate economy rate
death_overs = death_balls[eligible_bowlers] / 6
death_economy = (death_runs[eligible_bowlers] / death_overs).round(2)

#  Step 6: Combine into DataFrame
death_stats_df = pd.DataFrame({
    'Overs': death_overs,
    'Runs': death_runs[eligible_bowlers],
    'Economy': death_economy
}).sort_values('Economy')

# üìã Final Output
print(" Top MI Death Bowlers (min 30 overs):")
print(death_stats_df.head(10))
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
sns.barplot(
    x='Economy', 
    y=death_stats_df.head(10).index,
    data=death_stats_df.head(10), 
    palette='coolwarm_r'
)
plt.title(" Top 10 MI Death Bowlers (min 30 overs)")
plt.xlabel("Economy Rate")
plt.ylabel("Bowler")
plt.grid(axis='x', linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
```

### 6.Powerplay Bowlers (Overs 1‚Äì6)

```python
# Step 1: Filter Powerplay overs
powerplay_bowling = mi_bowling[mi_bowling['phase'] == 'Powerplay']

# Step 2: Filter legal deliveries
powerplay_legal = powerplay_bowling[~powerplay_bowling['extras_type'].isin(['wides', 'noballs'])]

# Step 3: Aggregate runs and balls
powerplay_runs = powerplay_bowling.groupby('bowler')['total_runs'].sum()
powerplay_balls = powerplay_legal.groupby('bowler').size()

# Step 4: Filter bowlers who bowled more than 90 balls (15 overs)
eligible_pp_bowlers = powerplay_balls[powerplay_balls > 90].index

# Step 5: Compute overs and economy
powerplay_overs = powerplay_balls[eligible_pp_bowlers] / 6
powerplay_economy = (powerplay_runs[eligible_pp_bowlers] / powerplay_overs).round(2)

# Step 6: Combine into DataFrame
powerplay_stats_df = pd.DataFrame({
    'Overs': powerplay_overs,
    'Runs': powerplay_runs[eligible_pp_bowlers],
    'Economy': powerplay_economy
}).sort_values('Economy')

# Final output
print(" Top MI Powerplay Bowlers (min 15 overs):")
print(powerplay_stats_df.head(10))
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(
    x='Economy', 
    y=powerplay_stats_df.head(10).index,
    data=powerplay_stats_df.head(10), 
    palette='Blues_d'
)
plt.title("Top 10 MI Powerplay Bowlers (min 15 overs)")
plt.xlabel("Economy Rate")
plt.ylabel("Bowler")
plt.grid(axis='x', linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
```

### 9.Player Consistency Scores
Create a custom consistency metric based on standard deviation and average performance:


```python
# Example for batsmen (MI only)
mi_batting = deliveries[deliveries['batting_team'] == 'Mumbai Indians']
player_runs_per_match = mi_batting.groupby(['match_id', 'batter'])['batsman_runs'].sum().reset_index()
player_stats = player_runs_per_match.groupby('batter')['batsman_runs'].agg(['mean', 'std', 'count'])
player_stats['consistency_score'] = (player_stats['mean'] / player_stats['std']) * player_stats['count']
top_consistent_batsmen = player_stats.sort_values('consistency_score', ascending=False).head(10)
print(top_consistent_batsmen[['mean', 'std', 'consistency_score']])
```
